```python
import os
class Product:
   def __init__(self, product_id, name, quantity, price):
       self.product_id = product_id
       self.name = name
       self.quantity = int(quantity)
       self.price = float(price)

   def __str__(self):
        return f"{self.name} (ID: {self.product_id}) | Stock: {self.quantity} | price: ${self.price:.2f}"

class InventoryManager:
    def __init__(self, file_name="inventory.txt"):
        self.file_name = file_name
        self.inventory = {}
        self.load_from_file()

    def load_from_file(self):
        if os.path.exists(self.file_name):
           with open(self.file_name, "r") as file:
               for line in file:
                   #strip newline and split by comma
                   pid, name, qty, price = line.strip().split(",")
                   self.inventory[pid] = Product(pid, name, qty, price)
    def save_to_file(self):
        with open(self.file_name,"w") as file:
            for pid, prod in self.inventory.items():
                file.write(f"{pid},{prod.name},{prod.quantity},{prod.price}\n" )
                
    def add_or_update_product(self, pid, name, qty, price):
        if pid in self.inventory:
            self.inventory[pid].quantity += qty  #Loop/update Logic
            print(f"Updated {name} quantity.")
        else:
            self.inventory[pid] = Product(pid, name, qty, price)
            print(f"added new product: {name}.")
            self.save_to_file()
    def check_low_stock(self, threshold=5):
        print("/n--LOW STOCK ALART--")
        low_stock_found = false
        for pid, prod in self.inventory.items():
            if prod.quantity <= threshold:  #Conditional statement
                print(f"{prod.name} is low on stock!! Only {prod.quantity} left.  ")
                low_stock_found = True

                if not low_stock_found:
                 print("All product stock levels are healthy.")

                 
#------Demonstration Run-------
if __name__=="__main__":
    manager= InventoryManager()


    #adding items (simulating Real-Time Inputs)
    p_id = input("Enter Product ID : ")
    p_name = input("Enter Product Name : ")
    p_qty = input("Enter Product Quantity : ")
    p_price = input("Enter Product Price : ")
    
    
    manager.add_or_update_product("101", "Wireless Mouse", 12, 25.00)

    manager.add_or_update_product("102", "HDMI Cable", 3, 10.50)

    manager.add_or_update_product("103", "Gaming Keyboard", 2, 80.00)

    manager.add_or_update_product(p_id, p_name, p_qty, p_price)

    
    manager.display_inventory()

    manager.check_low_stock()


    #passing the runtime user inputs to the manager
    manager.add_or_update_product(p_id, p_name, int(p_qty), float(p_price))

```
